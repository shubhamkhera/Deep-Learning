## Use Case: An HR Company has hired you as a Data Scientist. Your job as a consultant is to prepare and deploy a model that can predict the salary of the employee on the basis of the Years of Experience.

```python
import  numpy as np
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('03_Salary_Data.csv')
data.info()
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
#Looks like there is no null value. Further assi=uming that there should be less null values in Population data as well,
#so lets drop the rows with empty values.

data.dropna(inplace=True)
```

```python
data.duplicated().sum()
```

```python
#Lets drop duplicates because of same logic as of null value.
data.drop_duplicates(inplace=True)
```

```python
# Rules For Regression for ANN -- DL
# =======================================
# 1. Data should be complete.
# 2. Data should be strictly Numeric.
# 3. Features must be represented in the form of 2d np array.
# 4. Label must be represented in the form of 2d np array.
# 5. Normalizing Features and Label is Mandatory.

# Guidelines for Normalization:
# 1. Features must be NORMALIZED   ----  Always use StandardScaler or Robust Scaler
# 2. Labels must be NORMALIZED    ----- Always use MinMaxScaler with range (0,1)
#
#
```

```python
# Create Feature and Label array
features = data.iloc[:,[0]].values
label = data.iloc[:,[1]].values
```

```python
# Feature Standardization
from sklearn.preprocessing import StandardScaler
std_sc_features = StandardScaler()
features = std_sc_features.fit_transform(features)
```

```python
# Label Normalization
from sklearn.preprocessing import MinMaxScaler
MinMaxLabel = MinMaxScaler()
label = MinMaxLabel.fit_transform(label)
```

```python
# label[0].item() # ==> 39343.0
MinMaxLabel.inverse_transform(np.array([label[0]])).item()
```

```python
# TrainTestSplit
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features,
                                                    label,
                                                    test_size= 0.2,
                                                    random_state=10)
```

### Model Building and Training Process in ANN

```python
# Guidelines for NUMBER of nodes and NUMBER of Hidden Layer for Classification
# ============================================================================
# How many NEURONS to be initialized
# 1. Option1: First Hidden Layer   --->   No. of Units   ---> No. of Features
#             No. of Hidden Layers --->  No. of Features

# 2. Option2: First Hidden Layer   --->   No. of Units   ---> 3 * No. of Features
#             No. of Hidden Layers --->  No. of Features

# 3. Option1: First Hidden Layer   --->   No. of Units   ---> 1/3 * No. of Features
#             No. of Hidden Layers --->  No. of Features   

# Note: If still model don't perform well try different combinations and try to find the best fit.


# Guidelines for NUMBER of nodes and NUMBER of Hidden Layer for Regression
# ========================================================================
# Start with 40 units in each hidden layer and increase the same till 100 units with increament of 5 each time.
```

### Step1: Architecting the model -- ANN

```python
import tensorflow as tf
#from tensorflow.keras.layers import Dense

# Step 1.1: Create Sequential Model object

model = tf.keras.Sequential()

# Step 1.2: Create Dense Layer
## units = No. of Neurons in Hidden Layer
## activation = Which activation Function to Apply
## input_shape = (NoOfColsInFeatureArray)

model.add(tf.keras.layers.Dense( units = 100, activation = 'sigmoid', input_shape = (1,) ))
model.add(tf.keras.layers.Dense( units = 100, activation = 'sigmoid'))
#model.add(tf.keras.layers.Dense( units = 100, activation = 'sigmoid'))
#model.add(tf.keras.layers.Dense( units = 100, activation = 'sigmoid'))
model.add(tf.keras.layers.Dense( units = 1, activation = 'linear'))
```

```python
# Custom Metric Function ----> R2_keras_custom
def r2_keras_custom(y_true, y_pred):
    SS_res = tf.keras.backend.sum(tf.keras.backend.square(y_true - y_pred))
    SS_total = tf.keras.backend.sum(tf.keras.backend.square(y_true - tf.keras.backend.mean(y_true)))
    
    return (1 - (SS_res / (SS_total + tf.keras.backend.epsilon())))
```

### Step2: Compile the model -- ANN

```python
model.compile(optimizer="sgd" , loss="mean_squared_error" , metrics=[r2_keras_custom])
```

### Step3: Fit the model with Training Data -- ANN

```python
model.fit(X_train, y_train, epochs = 1000, validation_data=(X_test, y_test))
```

```python
# Score
model.evaluate(X_train, y_train)
```

```python
model.evaluate(X_test, y_test)
```

```python
# 2. Check the quality of the model
# For Regression quality is judged based on Test Score > CL
# SL = 0.05
# CL = 1 - SL = 0.95
# Since > CL therefore Model APPROVED !!! 
```

```python
# Deployment with Input Example:

# 1. Getting user input in the form of float
yearOfExperience = float(input("Enter Years of Experience: "))

# 2. Converting input to 2d array since the same format was used for training this model
yearOfExperience_Array = np.array([[yearOfExperience]])

# 3. Apply Standardization scince the same was done during the training of model
scaled_yearOfExperience = std_sc_features.transform(yearOfExperience_Array)

# 4. Predicting the salary
salary = model.predict(scaled_yearOfExperience)

# 5. Inverse Transform Label
actualSalary = MinMaxLabel.inverse_transform(salary)

print(f"Salary Predicted for {yearOfExperience} years of Experience is {actualSalary}")
```

## Early Stopping Using CustomCallBack

```python
# if you want to create a terminating condition which can extract generalized model with test score > CL, its not possible
# directly. You will need to create a Custom CALLBACKS

# val_r2_keras_custom > r2_keras_custom & val_r2_keras_custom > CL

class MyThresholdCallBack(tf.keras.callbacks.Callback):
    def __init__(self,cl):
        super(MyThresholdCallBack, self).__init__()
        self.cl = cl
    
    def on_epoch_end(self, epoch, logs=None):
        test_score = logs["val_r2_keras_custom"]
        train_score = logs["r2_keras_custom"]
        
        if test_score > train_score and test_score > self.cl and train_score > self.cl:
            self.model.stop_training = True
```

## Step1: Architecting the Model -- ANN

```python
model1 = tf.keras.Sequential()

model1.add(tf.keras.layers.Dense( units = 100, activation = 'sigmoid', input_shape = (1,) ))
model1.add(tf.keras.layers.Dense( units = 100, activation = 'sigmoid'))
model1.add(tf.keras.layers.Dense( units = 1, activation = 'linear'))
```

```python
# Custom Metric Function ----> R2_keras_custom
def r2_keras_custom(y_true, y_pred):
    SS_res = tf.keras.backend.sum(tf.keras.backend.square(y_true - y_pred))
    SS_total = tf.keras.backend.sum(tf.keras.backend.square(y_true - tf.keras.backend.mean(y_true)))
    
    return (1 - (SS_res / (SS_total + tf.keras.backend.epsilon())))
```

### Step2: Compile the model -- ANN

```python
model1.compile(optimizer="sgd" , loss="mean_squared_error" , metrics=[r2_keras_custom])
```

### Step3: Fit the model with Training Data -- ANN


```python
myR2ScoreMonitor = MyThresholdCallBack(cl=0.95)

model1.fit(X_train, y_train, epochs = 10000, validation_data=(X_test, y_test), callbacks=[myR2ScoreMonitor])
```

```python
# Score
model1.evaluate(X_train, y_train)
```

```python
model1.evaluate(X_test, y_test)
```
