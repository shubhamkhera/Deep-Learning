### Significance Level:
- **It defines how much %age(percentage) error is tolerable in the model such that the model is in comploiance with the SLA(Service Level Agreement)**


- SL: 0.05(5%)   
- Accuracy Required: 0.95(95%) 

### Confidence Level:
- **Minimum accuracy required i.e., 1 - SL. It is defined as the min accuracy expected from the model**

## Multi-Layer Neuron(Perceptron) {Artificial Neural Network}:

- A neuron(Perceptron) typically perform 2 Operations
    - Summation Operation
    - Activation operation
    
    
- **Input Layer: Consist of inputs only**
- **Hidden Layer: It consist of multiple neurons along with activation function**
- **Output Layer: It gives us the final predicted values**
    
    





### Loss Function/ Cost Function: 
- **Finding the Difference between the actual and the predicted values**
- **Loss = Y - Ypred**
- **Applicatble i.e., calculated for each record**


### Error Function:
- **It is summition of loss Finction for all n(number of records) divided by n(number of records)**




### The Loss/Cost/Error Function is only applicable for Supervised Learnings Further... :
- **Regression**
    - **Squared Error => LossFunction = (Y - Ypred)^2**
    - **Absolute Error => LossFunction = |Y - Ypred|**
    - **Mean Squared Error => ErrorFunction = (sum from i to n (Y - Ypred)^2) / n**
    - **Mean Absolute Error => ErrorFunction = (sum from i to n |Y - Ypred)| / n**

- **Classification**
    - **Cross Entropy**
    - **Ginni ENtropy**
